```python
%matplotlib inline
import cv2
import matplotlib.pyplot as plt
img = cv2.imread('Hemanvi.jpeg')
#   
```

```python
img = cv2.imread('Hemanvi.jpeg')
# crop the image
img = img[50:250,40:240]
#convert image to grayscale
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
#show image
plt.imshow(img_gray, cmap = 'gray')
```

```python
#convert the image inti a 25 * 25 array 
img_gray_small = cv2.resize(img_gray,(25,25))
plt.imshow(img_gray_small, cmap= 'gray')
```

```python
print(img_gray_small)
```

```python
# crop for the color image
img = cv2.imread('Hemanvi.jpeg')
img = img[50:250, 40:240,:]
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
print(img.shape)
```

```python
# The bottom-right 3 x 3 array of pixels can be obtained as follows
crop = img[-3:,-3:]
print(crop)
plt.imshow(crop)
```
